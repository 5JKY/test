```python
import numpy as np
import lammps_logfile
import matplotlib.pyplot as plt
```

### Energy minimization and MD at temp0=1.2

```python
log_min = lammps_logfile.File("log.min")
timestep = 0.001
Time_EM = log_min.get("Step", run_num = 0)*timestep
PotEng_EM = log_min.get("PotEng", run_num = 0)
KinEng_EM = log_min.get("KinEng", run_num = 0)
plt.plot(Time_EM, PotEng_EM)
plt.plot(Time_EM, KinEng_EM)
plt.grid()
```

```python
log0 = lammps_logfile.File("log.0")
Time0 = log0.get("Step")*timestep
Time0 -= Time0[0]
PotEng0 = log0.get("PotEng")
KinEng0 = log0.get("KinEng")
TotEng0 = log0.get("TotEng")
Temp0 = log0.get("Temp")
Press0 = log0.get("Press")

ave0_temp = np.average(Temp0[np.where(Time0 == 2)[0][0]:])
ave0_press = np.average(Press0[np.where(Time0 == 2)[0][0]:])
ave0_potential = np.average(PotEng0[np.where(Time0 == 2)[0][0]:])
```

```python
plt.plot(Time0, Temp0)
plt.title(f"tempreture {ave0_temp}")
plt.grid()
```

```python
plt.plot(Time0, KinEng0, label="kinitic")
plt.plot(Time0, PotEng0, label="potential")
plt.plot(Time0, TotEng0, label="total")
plt.title("energy plot")
plt.legend()
plt.grid()
```

```python
plt.plot(Time0, Press0)
plt.title(f"pressure {ave0_press}")
plt.grid()
```
