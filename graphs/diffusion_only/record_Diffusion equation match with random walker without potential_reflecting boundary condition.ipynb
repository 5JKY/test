```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
D = 0.01  # Diffusion coefficient
L = 1.0  # Length of the domain
hx = 0.01 # Spatial grid spacing
Nx = int(L / hx) + 1 # Number of spatial grid points 

ht = 0.005 # Time step size
```

```python
# Parameters --- ReRe
num_particles = 100000
num_steps = 2000  # Number of steps
positions_arr = L/4 * np.ones(num_particles, dtype=float)
step_size = hx/2   # Size of each step

# Generate random steps: (hx/2)^2=2D*(ht/4)
for i in np.arange(num_steps*4):
    positions_arr += np.random.choice([-step_size, step_size], size=positions_arr.size)
    lower_reflect_id = np.where(positions_arr < 0.0)
    positions_arr[lower_reflect_id] += 2*step_size
    # Use np.where() to find indices of values within the range
    upper_reflect_id = np.where(positions_arr > 1.0)
    positions_arr[upper_reflect_id] -= 2*step_size
    # Use np.where() to find indices of values within the range
    indices = np.where((positions_arr > 0.0) & (positions_arr < 1.0))
    # Select values within the range using the indices, filter outfielder 
    positions_arr = positions_arr[indices]
```

```python
unique_values, counts = np.unique(positions_arr, return_counts=True)
# concentrations in every spacial step
concentrations=counts/(2*step_size*num_particles)
```

```python
plt.plot(unique_values, concentrations)
plt.grid()
```

```python
# Parameters --ReAb
num_particles = 100000
num_steps = 8000  # Number of steps
positions_arr = L/4 * np.ones(num_particles)
step_size = hx/2   # Size of each step

# Generate random steps: (hx/2)^2=2D*(ht/4)
for i in np.arange(num_steps*4):
    positions_arr += np.random.choice([-step_size, step_size], size=positions_arr.size)
    reflect_id = np.where(positions_arr < 0.0+step_size/3)
    positions_arr[reflect_id] += 2*step_size
    # Use np.where() to find indices of values within the range
    absorb_id = np.where(positions_arr < L-step_size/4)
    # Select values within the range using the indices, filter outfielder 
    positions_arr = positions_arr[absorb_id]
```

```python
unique_values, counts = np.unique(positions_arr, return_counts=True)
```

```python
# Initial condition (delta function)
def initial_condition(x, x0):
    return np.where(np.abs(x - x0) < hx / 2, 1.0 / hx, 0.0)

# Initialize temperature array
u = np.zeros(Nx)

# Set initial condition (delta function at x = L/4)
x_values = np.linspace(0, L, Nx)
u[:] = initial_condition(x_values, L / 4)

# Implicit Euler time-stepping loop
for n in range(1,  num_steps+1):
    # Assemble coefficient matrix A
    A = np.diag(-2*np.ones(Nx-1)) + np.diag(np.ones(Nx-2), k=-1) + np.diag(np.ones(Nx-2), k=1)
    # Apply reflecting boundary condition
    A[0][1] = 2
    A *= D * ht / hx**2
    
    # Apply absorbing boundary condition
    u[-1] = 0
    
    # Solve the system Au^{n+1} = u^n
    u[0:-1] = np.linalg.solve(np.eye(Nx-1)-A, u[0:-1])
    
plt.plot(x_values, u, label=f"t={num_steps*ht:.2f} numerical solution")
plt.plot(unique_values, concentrations, label=f"t={num_steps*ht:.2f} random walker")

# Plot formatting
plt.xlabel('x')
plt.ylabel('concentration C(x,t)')
plt.title('numerical solution of diffusion equation and evoltion of random walker (without potential-ReAb) ')
plt.legend()
plt.grid(True)

plt.show()
```

```python
def calculate_r_squared(y_true, y_pred):
    """
    Calculate the coefficient of determination (R^2) for a given set of true and predicted values.
    """
    # Calculate the mean of the true values
    mean_y_true = np.mean(y_true)
    
    # Calculate the total sum of squares (TSS)
    tss = np.sum((y_true - mean_y_true)**2)
    
    # Calculate the residual sum of squares (RSS)
    rss = np.sum((y_true - y_pred)**2)
    
    # Calculate R^2
    r_squared = 1 - (rss / tss)
    
    return r_squared
```

```python
y_true = u[1:-1]  # True values
y_pred = concentrations # Predicted values

# Calculate R^2
r_squared = calculate_r_squared(y_true, y_pred)
print("R^2:", r_squared)
```
